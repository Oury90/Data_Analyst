```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df.dropna(subset=['price_paid'], inplace=True)
df
```

```python
df.drop(['id', 'first_name', 'last_name', 'email', 'ip_address'], axis=1, inplace=True)
df
```

```python
df.price_paid = df.price_paid.apply(lambda x: x.replace("$", ""))
df.price_paid = df.price_paid.astype(float)
df
```

```python
somari = df.groupby('country')["price_paid"].sum().plot.bar(rot=45, figsize=[20, 10], legend=True)
```
